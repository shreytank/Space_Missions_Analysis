```python
import math
import decimal
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
import decimal 
from decimal import *
```

```python
def D(x,u):
    e=math.e
    n1=Decimal(pow(e,-u))
    n2=(pow(u,x))
    d=(math.factorial(x))
    p=Decimal((n1*n2)/d)
    return p
```

```python
data=pd.read_csv("C:/Users/shrey/Downloads/IBM/Chisquare.csv")
data
```

```python
p=[]
success=np.array(data['Success'])

for x in range(len(data)):
    p.append(1-round(D(success[x],3),3))

for x in range(len(data)):
    print(p[x])
```

```python
data['Probability']=p
data
```

```python
plt.figure(figsize=(15,10))
g=sns.barplot(y='Countries', x='Probability', data=data)
```
