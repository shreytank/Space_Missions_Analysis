```python
import math
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data1=pd.read_csv("C:/Users/shrey/Downloads/IBM/Chisquare.csv")
data1
```

```python
avg=sum(data1['Success'])/len(data1)
obs=np.array(data1['Success'])
res=[]
for x in range(len(data1)):
    res.append(obs[x]-avg)

res2=[]    
for x in range(len(data1)):
    res2.append(res[x]**2)
    
final=[]
for x in range(len(data1)):
    final.append(res2[x]/avg)
    
ans=sum(final)
print("chi-square statistic:",ans)   
```
